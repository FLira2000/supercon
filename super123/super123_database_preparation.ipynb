```python
import pandas as pd
import numpy as np
```

```python
allMaterials = pd.read_csv('allElements_unique_m.csv')
```

```python
allMaterials
```

```python
super123materials = pd.read_csv('super123_materials.csv')
```

```python
super123materials
```

```python
newMaterialsList = pd.DataFrame()
```

```python
for i, row in super123materials.iterrows():
    temp = allMaterials.loc[allMaterials['material'] == row['material']]
    material = ''
    hghr = 0
    if len(temp) > 1:
        for j, rw in temp.iterrows():
            if float(rw['critical_temp']) > hghr:
                hghr = float(rw['critical_temp'])
                material = rw['material']
        temp = temp.loc[temp['material'] == material]
        print(temp)
        temp = temp.loc[temp['critical_temp'] == hghr]
        print(temp)
    
    newMaterialsList = newMaterialsList.append(temp)
```
