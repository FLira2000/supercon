```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../super123Database_final', index_col=[0])
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0.1', axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df_only_num = df.drop('material', axis=1)
```

```python
df_only_num.columns
```

```python
df_only_num.shape
```

```python
scaled_df = preprocessing.scale(df_only_num)
```

```python
pca = PCA()
```

```python
pca.fit(scaled_df)
```

```python
scaled_df.shape
```

```python
pca_data = pca.transform(scaled_df)
```

```python
import numpy as np
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
```

```python
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
pca_df = pd.DataFrame(pca_data,index=[*df['material']], columns=labels)
```

```python
pca_df.tail()
```

NÃ£o pudemos reduzir dimensionalmente em poucos componentes, logo, plotar o grÃ¡fico seria desumano.

```python
pca_df.shape
```

```python
pca_df.to_csv('../pca_df.csv', index=False)
```

```python
sm_percent = 0
dim_reduc_pcs = []
for pc in per_var:
    sm_percent+=pc
    dim_reduc_pcs.append(pc)
    if(sm_percent > 98):
        break
```

```python
print(len(dim_reduc_pcs))
```

We can representate the data with 23 principal components;

```python
print(sm_percent)
```

With 98.3% of certain in representing the original data.

```python
reduce_labels = ['PC' + str(x) for x in range(1, len(dim_reduc_pcs)+1)]
print(reduce_labels)
```

```python
reduced_pca_df = pca_df[[*reduce_labels]]
```

```python
reduced_pca_df.shape
```

```python
reduced_pca_df.to_csv('../reduced_pca_pf.csv')
```

```python
plt.bar(x=range(1,len(dim_reduc_pcs)+1), height=dim_reduc_pcs, tick_label=labels[:len(dim_reduc_pcs)])
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

This graphic shows the principal components needed to represent, with 98% of precision, our dataset.
