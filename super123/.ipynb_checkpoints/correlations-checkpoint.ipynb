```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('allElements_unique_m.csv')
```

```python
df
```

```python
featureList = df.columns
```

```python
featureList = featureList.drop('Unnamed: 0').drop('material')
```

```python
featureList
```

```python
correlationList = []
for feature in featureList:
    if feature == 'critical_temp':
        continue
    correlationList.append( [ feature, df['critical_temp'].corr(df[feature], method='spearman') ] )
```

```python
correlationList
```

```python
goodCorrelationList = []
for i in correlationList:
    print(i[1])
    if i[1] > 0.85 or i[1] < -0.85:
        goodCorrelationList.append( i )        
```
