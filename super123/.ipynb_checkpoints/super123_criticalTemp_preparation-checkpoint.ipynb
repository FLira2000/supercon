```python
elements = pd.read_csv("unique_materials.csv")
```

```python
elements
```

```python
elements = elements[["material", "critical_temp"]]
```

```python
elements
```

```python
def meanTemperature(temp):
    mean = 0
    if str(type(temp)) == "<class 'pandas.core.frame.DataFrame'>":
        temperatureList = temp['critical_temp']
        temp = np.array(temperatureList)
        mean = sum(temp)/len(temperatureList)
    elif str(type(temp)) == "<class 'numpy.ndarray'>":
        print('arr')
        mean = sum(temp)/len(temp)

    return mean
```

```python
lst = elements['material'].values.tolist()
shadowLst = []
for i, row in df.iterrows():
    if row['material'] in lst:
        temp = elements.loc[elements['material'] == row['material']]
        temp2 = np.array(temp)
        if (len(temp2)) > 1:
            temp = meanTemperature(temp)
            shadowLst.append(float(temp))
        else:
            shadowLst.append(float(temp['critical_temp']))
            
```

```python
df['critical_temp'] = shadowLst
```

```python
import math
cnt = 0
for temp in df['critical_temp'].unique():
    if math.isnan(temp) == True:
        cnt += 1
print(cnt)
```
