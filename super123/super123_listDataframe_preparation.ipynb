```python
import pandas as pd
```

```python
file = open('super123.txt')
```

```python
file = file.read()
```

```python
lst = []
material = []
for letter in file:
    material.append(letter)
    if letter == '\n':
        material.pop()
        material = ''.join(material)
        print(material)
        lst.append(material)
        material = []
```

```python
df = pd.DataFrame({'material': lst})
```

```python
df
```

```python
df = df.dropna()
```
