```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
from math import sqrt
```

```python
df = pd.read_csv('../super123Database_final')
```

```python
df = df.drop('Unnamed: 0', axis=1).drop('Unnamed: 0.1', axis=1)
```

```python
df.columns
```

```python
y = df['critical_temp']
x = df.select_dtypes(exclude=['object']).drop('critical_temp', axis=1)
```

```python
x
```

```python
train_X, test_X, train_y, test_y = train_test_split(x.values, y.values, test_size=0.2)
df_imputer = SimpleImputer()
train_X = df_imputer.fit_transform(train_X)
test_X = df_imputer.transform(test_X)
```

```python
#rede neural com activator relu de duas camadas
```

```python
neuralNet_relu2 = MLPRegressor(hidden_layer_sizes = (60, 30), activation = 'relu', solver="lbfgs")
```

```python
neuralNet_relu2.fit(train_X, train_y)
```

```python
predictions_relu2 = 0
```

```python
avg_relu2 = 0
for i in range(1, 30):
    predictions_relu2 = neuralNet_relu2.predict(test_X)
    avg_relu2 += sqrt(mean_absolute_error( predictions_relu2, test_y ))
print(str(avg_relu2/30))
```
