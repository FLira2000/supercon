```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
from math import sqrt
```

```python
df = pd.read_csv('../super123Database_final')
```

```python
df = df.drop('Unnamed: 0', axis=1).drop('Unnamed: 0.1', axis=1)
```

```python
df.columns
```

```python
y = df['critical_temp']
x = df.select_dtypes(exclude=['object']).drop('critical_temp', axis=1)
```

```python
x
```

```python
train_X, test_X, train_y, test_y = train_test_split(x.values, y.values, test_size=0.2)
df_imputer = SimpleImputer()
train_X = df_imputer.fit_transform(train_X)
test_X = df_imputer.transform(test_X)
```

```python
#rede neural com activator tanh de uma camada
```

```python
neuralNet_tahn = MLPRegressor(hidden_layer_sizes = 90, activation = 'tanh', max_iter=2000)
```

```python
neuralNet_tahn.fit(train_X, train_y)
```

```python
avg_tahn = 0
for i in range(1, 30):
    predictions_tahn = neuralNet_tahn.predict(test_X)
    avg_tahn += sqrt(mean_absolute_error( predictions_tahn, test_y ))
print(str(avg_tahn/30))
```
